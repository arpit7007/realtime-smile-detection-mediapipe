```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
import joblib
```

```python
data = pd.read_csv("../data/smile_dataset.csv")
```

```python
x= data[["mouth_height"]]
y=data["smile"]
```

```python
model = LogisticRegression()
model.fit(x,y)

joblib.dump(model, "../models//smile_model.pkl")
```
